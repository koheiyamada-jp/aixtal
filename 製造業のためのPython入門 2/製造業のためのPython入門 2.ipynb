リスト配列同士の掛け算であるmulti_listはエラーメッセージのように実行できません。  
リストの掛け算という機能は組み込み関数にはありません。

**Numpy.ndarray**
- Numpyをインポートする必要がある
- 同じ型しか格納できない
- 多次元配列を表現できる
- 数値計算のためのメソッド・関数が豊富で、高速な演算が可能

````python
import numpy as np # 既に上でNumpyをインポートしているため本当は不要

A = np.array([[1, 1],    # 紹介したリストのNumpy配列変換
              [1, 1]])
B = np.array([[1, 2],
              [3, 4]])

add_array = A + B
print(add_array)

print("----------")

multi_array = A * B
print(multi_array)

print("----------")

dot_array = np.dot(A, B)
print(dot_array)
````

- add_arrayはNumpyの配列（行列）の要素同士の足し算  
- multi_arrayはNumpyの配列（行列）の要素同士の掛け算  
- dot_arrayはNumpyの配列（行列）同士の掛け算、いわゆる内積  
であり、組み込み関数ではできない機能です。  

ここまででリストの配列とNumpyのndarrayのごく一部の機能を比較して、その違いを紹介しました。  
まだまだNumpyには多くの実践的な機能を備えています。  
全てを紹介できませんが、実際の解析やその前処理を行う中でNumpyの有能性を実感していきましょう。

2次元配列にしたbですが、axisを指定せずにappendすると第一引数および第二引数の配列が一次元に平坦化されて結合します。

次元を保ったまま配列を追加するには引数axisを指定します。  
axis=0にすると新たな行として追加（縦に連結）、axis=1にすると新たな列として追加（横に連結）されます。  
存在しない次元（軸）を指定するとエラーになります。

````python
np.append(b,[[12, 13, 14, 15]],axis = 0)
````

````python
np.append(b,[[12, 13, 14, 15]],axis = 1) # 3行×4列の2次元配列bと1行×4列の2次元配列を横に結合することはできない
````

````python
np.append(b,[12, 13, 14, 15], axis = 0) # 結合元と結合先の次元数が一致していないときはエラーが発生する
````

### グラフの描画

製造業の皆さんが取得されるデータをグラフとして可視化する方法を紹介します。

グラフを出力する大体の流れは  
- matplotlib.pyplotのインポート
- plt."グラフの種類を選択"( データなど )
- plt."グラフのスタイルオプションを選択"
- plt.show()でグラフを出力
- plt.savefig("ファイル名.png")で画像を保存  

となります。

````python
plt.plot(                   # plotは折れ線グラフ
         [1,2,3,4],         # xデータ
         [1,4,9,16],        # yデータ
         color="red",       # 赤色
         linestyle="solid", # 実線、"None"にすると非表示
         linewidth=2,       # ライン幅は2pt
         marker="d",        # マーカーの形状
         markersize=8,      # マーカーサイズ
         markerfacecolor="silver",  # マーカーの塗りつぶし色
         markeredgewidth=1,         # マーカーの端の幅
         markeredgecolor="black"    # マーカーの端の色
        )       
plt.xlabel("x-label")       # x軸のラベル
plt.ylabel("y-label")       # y軸のラベル
plt.axis([0,                # x軸の最小値
          5,                # x軸の最大値
          0,                # y軸の最小値
          20])              # y軸の最大値
plt.grid()                  # グリッド線

plt.show()                  # グラフの出力
````

colorとlinestyleのバリエーションが多数あるので必要であれば調べてみましょう

### オブジェクトの挿入

````python
plt.plot(range(10)) # y値を0～9にしてラインをプロット

plt.text(0,         # テキストの左端：x軸の0の位置
         8.5,       # テキストのベースライン：y軸の8.5の位置
         "test",    # 表示するテキスト
         size=30    # フォントサイズ[pt]
         )

plt.text(9.8,
         5.0,
         "python",
         size=20,
         ha="left",                    # テキストの左端を基準
         va="bottom",                  # テキストの底を基準
         bbox=dict(boxstyle="circle",  # ボックスの形状
                   facecolor="pink",   # 塗りつぶし色
                   edgecolor="red",    # 縁の色
                   linestyle="solid",  # 縁の線のスタイル
                   linewidth=3         # 縁の線幅
                  )
         )

plt.text(2.0,
         3.5,
         "Machine\nLearning",           # テキストの改行する位置に"\n"を挿入
         size=20,
         rotation=33,                   # 回転
         bbox=dict(boxstyle="round",    # ボックスの形状
                   facecolor="orange",  # 塗りつぶし色
                   edgecolor="red",     # 縁の色
                   linestyle="solid",   # 縁の線のスタイル
                   linewidth=1          # 縁の線幅
                  )
         )

plt.annotate("Notice!",                             # 注釈テキスト
             xy=(4,2),                              # 矢印の先端の位置
             xytext=(5,3.5),                        # 注釈テキストの開始位置
             size=20,
             arrowprops=dict(facecolor="orange",  
                             width=1,               # 矢印の線幅
                             headwidth=10,          # 矢印の大きさ
                             headlength=15,         # 矢印の長さ
                             linewidth=1   ,        # エッジラインの太さ
                             edgecolor="black",     # エッジラインの色
                             shrink=0.05            # 矢印の先端を離す
                             )
             )

plt.grid()
plt.show()
````

グラフ1つの場合：  
FigureというキャンバスにAxesというグラフが配置し、その中の構成物としてAxisがあります。

グラフ3つの場合：  
FigureというキャンバスにAxesというグラフが3つ配置しています。

このFigureとAxesの関係性を考慮すると以下のようにコードを記述できます。

````python
names=["groupA","groupB","groupC"]
values=[1,10,100]

plt.figure(1,              # FigureオブジェクトのID
           figsize=(9,3),  # figureのサイズ
           edgecolor="red", # エッジラインの色
           linewidth=5      # エッジラインの幅
          )
plt.suptitle("Figure Title", fontsize=16,y=1.2)

ax1 = plt.subplot(131)
ax1.bar(names,values)      # 棒グラフ
ax1.set_title("Axe1")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2 = plt.subplot(132)
ax2.plot(names,values)     # 折れ線グラフ
ax2.set_title("Axe2")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax3 = plt.subplot(133)
ax3.scatter(names,values)  # 散布図
ax3.set_title("Axe3")
ax3.set_xlabel("x")
ax3.set_ylabel("y")


plt.subplots_adjust(wspace=0.5,  # グラフ間のスペース
                    top=1.0)     # Figureとサブプロットエリアの上スペース

plt.show()
````

上記のコードを他の方法（配列形式）で記述します。

````python
names=["groupA","groupB","groupC"]
values=[1,10,100]

fig, ax =plt.subplots(1,3,               # 1×3のグリッド
                      figsize=(9,3),
                      edgecolor="red",
                      linewidth=5
                      )
fig.suptitle("Figure Title", fontsize=16,y=1.1)

ax[0].bar(names,values)
ax[0].set(title="Axe0",
            xlabel="x",
            ylabel="y"
           )

ax[1].plot(names,values)
ax[1].set(title="Axe1",
            xlabel="x",
            ylabel="y"
           )

ax[2].scatter(names,values)
ax[2].set(title="Axe2",
            xlabel="x",
            ylabel="y"
           )

plt.show()
````

4つのグラフを2×2として表示し、全てのスペースを埋めるようにします。

````python
x = np.linspace(0,2*np.pi,400)
y1 = np.sin(x)
y2 = np.sin(x**2)
y3 = np.cos(x/2)
y4 = np.cos(x**2)

fig, ax =plt.subplots(2,2,          # 2×2のグリッド
                      sharex=True,  # x軸を共有
                      sharey=True   # y軸を共有
                      )

fig.suptitle("Figure Title", fontsize=16,y=1)

ax[0,0].scatter(x,y1)

ax[0,1].scatter(x,y2)

ax[1,0].scatter(x,y3)

ax[1,1].scatter(x,y4)

fig.subplots_adjust(hspace=0,   # 縦方向のスペースを0にする
                    wspace=0    # 横方向のスペースを0にする
                   )

plt.show()
````

他の3Dグラフを表示します。

````python
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d

def func(x,y):
    return x**2 + y**2

x1 = np.arange(-3,3,0.25)  # x_1軸を生成
x2 = np.arange(-3,3,0.25)  # x_2軸を生成
X,Y = np.meshgrid(x1,x2)   # 2次元の格子座標を生成
Z = -func(X,Y)              # 関数funcに格子点の各X,Y座標を代入してZを算出

fig = plt.figure()                    
ax = axes3d.Axes3D(fig)                # Axes3Dオブジェクトを配置
ax.set_xlabel("x1")                    # x1の軸ラベル
ax.set_ylabel("x2")                    # x2の軸ラベル
ax.set_zlabel("f(x,y)", size=15)   # f(x,y)の軸ラベル
ax.plot_wireframe(X,Y,Z)               # x1,x2,f(x1,x2)の曲線をプロット

plt.show()
````

np.meshgridについて説明します。

目的を整理しましょう。  

2次元平面を格子面として、各格子点のx座標とy座標の組み合わせをfunc(x,y)に代入し、  
各格子点での関数funcに対するz値を取得したいのです。

格子面を作成するメソッドがnp.meshgridです。

````python
x = np.arange(1,6,1)
x
````

````python
y = np.arange(1,6,1)
y
````

````python
X,Y = np.meshgrid(x,y)
print(X)
print("----------")
print(Y)
````

変数Xには格子点のx座標、変数Yには格子点のy座標の情報が入ります。
